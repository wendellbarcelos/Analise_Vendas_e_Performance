#### Será necessário remover os pronomes, é visivel que tem clientes com o genero trocado, sendo necessário alterar o genero e mudar o valor de F e M por feminino e masculino

```python
# Visualizando os nomes que contém pronomes ( Dr, Dra, Sr, Sra, Srta) - necessário tratamento
resultado = clientes[clientes['Nome'].str.contains(r'\.', na=False)]
resultado
```

```python
# Alterando os valores usando um regex, removendo os pronomes dos nomes
clientes['Nome'] = clientes['Nome'].str.replace(r'^.*?\.\s*', '', regex=True)
clientes.sample(10)
```
